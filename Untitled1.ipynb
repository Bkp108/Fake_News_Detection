### Step 1: Data Understanding


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
```

> #### Load Data

```python
file_path = "fraud.csv"  # Update this path as needed
try:
    df = pd.read_csv("/content/Fraud.csv")
    print(f"✅ Data loaded successfully. Shape: {df.shape}")
except FileNotFoundError:
    print("❌ ERROR: File not found. Please check the path.")
    raise
```

### Step 2: DATA CLEANING + FEATURE ENGINEERING

```python
# Drop ID columns if present
for col in ['nameOrig', 'nameDest', 'step', 'isFlaggedFraud']:
    if col in df.columns:
        df.drop(col, axis=1, inplace=True)

# Encode 'type' column
if 'type' in df.columns:
    df = pd.get_dummies(df, columns=['type'], drop_first=True)

# Create derived features if balance columns exist
balance_cols = ['oldbalanceOrg', 'newbalanceOrig', 'oldbalanceDest', 'newbalanceDest']
if all(col in df.columns for col in balance_cols):
    df['diffOrig'] = df['oldbalanceOrg'] - df['newbalanceOrig']
    df['diffDest'] = df['newbalanceDest'] - df['oldbalanceDest']
    df['isOrigBalanceZero'] = (df['oldbalanceOrg'] == 0).astype(int)
    df['isDestBalanceZero'] = (df['oldbalanceDest'] == 0).astype(int)

# Drop rows where target is missing
if df['isFraud'].isnull().sum() > 0:
    print(f"⚠️ Dropping {df['isFraud'].isnull().sum()} rows with missing target label.")
    df = df[df['isFraud'].notna()]

print(f"✅ Final dataset shape after cleaning: {df.shape}")
```

### STEP 3: EXPLORATORY DATA ANALYSIS

```python
plt.rcParams['figure.figsize'] = (10, 6)
sns.countplot(x='isFraud', data=df)
plt.title("Fraud vs Not Fraud Distribution")
plt.savefig('fraud_distribution.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
if 'amount' in df.columns:
    sample_df = df.sample(n=50000, random_state=42) if len(df) > 100000 else df
    sns.histplot(sample_df[sample_df['isFraud'] == 1]['amount'], bins=50, label='Fraud', color='red', alpha=0.6)
    sns.histplot(sample_df[sample_df['isFraud'] == 0]['amount'], bins=50, label='Not Fraud', color='blue', alpha=0.6)
    plt.title('Amount Distribution by Class')
    plt.legend()
    plt.yscale('log')
    plt.savefig('amount_distribution.png', dpi=300, bbox_inches='tight')
    plt.show()
```

### STEP 4: MODEL BUILDING

```python
X = df.drop('isFraud', axis=1)
y = df['isFraud']

if len(df) > 1000000:
    print("⚠️ Large dataset detected. Using 500k sample for training.")
    df_sample = df.sample(n=500000, random_state=42, stratify=y)
    X = df_sample.drop('isFraud', axis=1)
    y = df_sample['isFraud']

# Final check for NaN in target
y = y.fillna(0)  

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

# Random Forest Model with tuned params
rf_model = RandomForestClassifier(
    n_estimators=200,
    max_depth=15,
    min_samples_split=10,
    class_weight='balanced',
    n_jobs=-1,
    random_state=42
)
rf_model.fit(X_train, y_train)

# Predictions
y_pred_rf = rf_model.predict(X_test)
y_probs_rf = rf_model.predict_proba(X_test)[:, 1]
```

### STEP 5: EVALUATION

```python
print("\nClassification Report:\n", classification_report(y_test, y_pred_rf))
print("\nROC AUC Score:", roc_auc_score(y_test, y_probs_rf))
```

```python

# Confusion Matrix
plt.figure()
sns.heatmap(confusion_matrix(y_test, y_pred_rf), annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix")
plt.savefig('confusion_matrix.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python

# ROC Curve
fpr, tpr, _ = roc_curve(y_test, y_probs_rf)
plt.figure()
plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {roc_auc_score(y_test, y_probs_rf):.3f})")
plt.plot([0, 1], [0, 1], linestyle='--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.savefig('roc_curve.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python

# Feature Importance
feat_imp = pd.Series(rf_model.feature_importances_, index=X.columns).sort_values(ascending=False)
feat_imp.head(15).plot(kind='barh')
plt.title("Top 15 Feature Importances")
plt.tight_layout()
plt.savefig('feature_importance.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier

models = {
    "RandomForest": RandomForestClassifier(n_estimators=200, max_depth=15, class_weight='balanced', n_jobs=-1, random_state=42),
    "LogisticRegression": LogisticRegression(class_weight='balanced', max_iter=1000, solver='liblinear'),
    "DecisionTree": DecisionTreeClassifier(max_depth=15, class_weight='balanced', random_state=42),
    # "XGBoost": XGBClassifier(n_estimators=100, scale_pos_weight=20, use_label_encoder=False, eval_metric='logloss', random_state=42)
}

results = {}
```

```python
for name, model in models.items():
    print(f"\n🚀 Training {name}...")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_probs = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else model.decision_function(X_test)

    print(f"\n📈 Evaluation: {name}")
    print(classification_report(y_test, y_pred))
    roc_score = roc_auc_score(y_test, y_probs)
    print(f"ROC AUC Score: {roc_score:.4f}")
    results[name] = roc_score

    # Confusion Matrix
    plt.figure()
    sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt="d", cmap="Oranges")
    plt.title(f"{name} - Confusion Matrix")
    plt.savefig(f'{name.lower()}_confusion_matrix.png', dpi=300, bbox_inches='tight')
    plt.show()

    # ROC Curve
    fpr, tpr, _ = roc_curve(y_test, y_probs)
    plt.figure()
    plt.plot(fpr, tpr, label=f"{name} (AUC = {roc_score:.3f})")
    plt.plot([0, 1], [0, 1], linestyle='--')
    plt.title(f"{name} - ROC Curve")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.legend()
    plt.savefig(f'{name.lower()}_roc_curve.png', dpi=300, bbox_inches='tight')
    plt.show()
```

```python
print("\n🔍 Final ROC AUC Scores:")
for name, score in results.items():
    print(f"{name}: {score:.4f}")
```
